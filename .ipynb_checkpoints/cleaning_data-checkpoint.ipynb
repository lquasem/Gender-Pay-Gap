```python
 # Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
path = "data/Median_weekly_earnings_of_full-time_wage_and_salary_workers_by selected characteristics.csv"
```

```python
# Use encoding to read the CSV in pandas
table =pd.read_csv(path, encoding="UTF-8")
```

```python
 # Preview of the DataFrame
table.head(10)
```

```python
#Rename the column names
table = table.rename(columns ={'HOUSEHOLD DATA\nANNUAL AVERAGES\n37. Median weekly earnings of full-time wage and ': 'By race',
                               'Column1': 'Number of Workers (in 1000) in 2017',
                               'Column2': 'Number of Workers (in 1000) in 2018',
                               'Column3': 'Median weekly earnings in 2017',
                               'Column4': 'Number of Workers (in 1000) in 2018'})
table.head()
```

```python
# Delete NA and age range rows by index
# table = table.drop(table.index[range(0,4)], axis=0)
table = table.drop(table.index[[0,1,2,3,4,6,8,9,10,12,13,14,15,19,23,27,31]])
table
```

```python
list(table.columns) 
```

```python
#Delete  2nd column that doesnt contain necessary data
table = table.drop(columns= ['HOUSEHOLD DATA\nANNUAL AVERAGES\n37. Median weekly earnings of full-time wage an.1'])
table
```

```python
# Identify incomplete rows
table.count()
```

```python
gender_and_race = table.reset_index(drop=True)
gender_and_race
```

```python
gender_and_race.to_csv(r'data\gender_and_race.csv', index=False)
```
